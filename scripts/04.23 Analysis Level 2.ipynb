```python
import pandas as pd
```

```python
pm_inv = pd.read_csv('../data/output csv/pm_inv_uses.csv')
```

```python
pm_inv = pm_inv.dropna(subset = ['air','heavy_metals','chlor_solv','lead(pb)','pesticides','petrol'], how = 'all')
```

```python
sites = pd.read_csv('../data/output csv/sites_df2.csv')
```

```python
pm_inv = pm_inv.fillna(0)
sites = sites.fillna(0)
```

```python
pm_inv = pm_inv.rename(columns = {'heavy_metals':'HM','chlor_solv':'chlor','lead(pb)':'Pb','pesticides':'pest'})
```

```python
pm_inv.head()
```

```python
site_suggest = pm_inv.assign(ag = '', auto = '', paint = '', dryclean = '', golf = '', industrial = '', parking = '', rr = '', road = '', residential = '', utility = '', fire = '')
```

```python
site_suggest
```

```python
site_suggest.columns
```

```python
sites
```

```python
for row_value, row_index in site_suggest.iterrows():
    
    if row_index['HM'] != 0 and row_index['pest'] != 0:
        site_suggest.at[row_value,'ag'] = 'all'
    elif row_index['HM'] != 0 or row_index['pest'] != 0:
        site_suggest.at[row_value,'ag'] = '1+'
        
    if row_index['chlor'] != 0 and row_index['petrol'] != 0:
        site_suggest.at[row_value,'auto'] = 'all'
    elif row_index['chlor'] != 0 or row_index['petrol'] != 0:
        site_suggest.at[row_value,'auto'] = '1+'
        
    if row_index['Pb'] != 0:
        site_suggest.at[row_value,'paint'] = 'all'
    
    if row_index['chlor'] != 0:
        site_suggest.at[row_value,'dryclean'] = 'all'
    
    if row_index['pest'] != 0:
        site_suggest.at[row_value,'golf'] = 'all'
        site_suggest.at[row_value,'utility'] = 'all'

    if row_index['air'] != 0 and row_index['chlor'] != 0 and row_index['HM'] != 0:
        site_suggest.at[row_value,'industrial'] = 'all'
    elif row_index['air'] != 0 or row_index['chlor'] != 0 or row_index['HM'] != 0:    
        site_suggest.at[row_value,'industrial'] = '1+'
        
    if row_index['petrol'] != 0:
        site_suggest.at[row_value,'parking'] = 'all'
    
    if row_index['air'] != 0 and row_index['chlor'] != 0 and row_index['pest'] != 0 and row_index['petrol'] != 0:
        site_suggest.at[row_value,'rr'] = 'all'
    elif row_index['air'] != 0 or row_index['chlor'] != 0 or row_index['pest'] != 0 or row_index['petrol'] != 0:
        site_suggest.at[row_value,'rr'] = '1+'
        
    if row_index['Pb'] != 0 and row_index['petrol'] != 0:
        site_suggest.at[row_value,'road'] = 'all'
    elif row_index['Pb'] != 0 or row_index['petrol'] != 0:
        site_suggest.at[row_value,'road'] = '1+'
    
    if row_index['air'] != 0 and row_index['pest'] != 0 and row_index['petrol'] != 0:
        site_suggest.at[row_value,'residential'] = 'all'
    elif row_index['air'] != 0 or row_index['pest'] != 0 or row_index['petrol'] != 0:
        site_suggest.at[row_value,'residential'] = '1+'    
        
    if row_index['air'] != 0:
        site_suggest.at[row_value,'fire'] = 'all'
```
